# Forced Oscillation

```python
# TODO: solve the IVP for a forced oscillation
def forced_oscillation(t, y, m, k, C, F0, wf):
    """
    The derivate function for a foced oscillation

    :param t: the time
    :param y: the initial condition y
    :param m: the mass
    :param k: the spring constant
    :param C: the damping coefficient
    :param F: the amplitude of the force
    :param wf: the frequency of the force
    """
    yderive = np.zeros(len(y))
    yderive[0] = y[1]
    yderive[1] = (-k*y[0] - C*y[1] + F0*np.cos(wf*t))/m

    return yderive
```

```python
from cProfile import label


m = 1
y0 = np.array([1, 0])
k = 1
t_span = (0, 30)
t_eval = np.linspace(0, 30, 50)
F0 = 0.5
wf = 0.8
l = 0.2
sol = solver.solve_ivp(forced_oscillation, t_span, y0, method="RK4", t_eval=t_eval, args=(m, k, l/2, F0, wf))
plt.plot(t_eval,sol[0],label=r"$\lambda=$"+str(l))
```
